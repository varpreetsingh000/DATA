Ans There are three different methods for creating identical 2D arrays in NumPy, followed by the code and output for each method:

Method 1: Using np.array()
We can manually define the 2D array by passing a list of lists to np.array().

```python
import numpy as np

# Method 1: Using np.array()
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array using np.array():\n", array1)
```

Method 2: Using np.ones() (or other similar functions like np.zeros() or np.full())
This method fills the array with a constant value (in this case, we will modify it to match the values of the target array).

```python
# Method 2: Using np.ones() and modifying the array
array2 = np.ones((2, 3))  # Creates a 2x3 array of ones
array2[0] = [1, 2, 3]     # Manually set first row
array2[1] = [4, 5, 6]     # Manually set second row
print("\nArray using np.ones() and modification:\n", array2)
```

Method 3: Using np.arange() with reshape()
This method generates a 1D array and reshapes it into a 2D array.

```python
# Method 3: Using np.arange() and reshape()
array3 = np.arange(1, 7).reshape(2, 3)
print("\nArray using np.arange() and reshape():\n", array3)
```

Q3
Ans Difference between np.array(), np.asarray(), and np.asanyarray():

np.array():

This function always creates a new array, regardless of whether the input is already an array-like structure (like a list or another NumPy array).
It copies the data into a new array unless explicitly told not to (copy=False), which ensures any changes to the original data don’t affect the new array.
It can take additional parameters like specifying the data type (dtype) or memory layout (order).

```python
#example
a = [1, 2, 3]
np_array = np.array(a)  # Creates a new NumPy array
```

np.asarray():

This function converts the input into a NumPy array, but it does not make a copy if the input is already a NumPy array. It only copies data when the input is not a NumPy array or if the data type (dtype) needs to change.
It's more efficient than np.array() when working with existing arrays since it avoids unnecessary copying.

```python
#example
arr = np.array([1, 2, 3])
np_asarray = np.asarray(arr)  # No copy is made if input is already an array
```

np.asanyarray():

Similar to np.asarray(), but it allows the creation of subclasses of arrays (like matrix, masked_array).
It doesn’t force the input into a base ndarray type, so if you pass an array of a subclass (like matrix), the result will retain the subclass type.

```python
#example
from numpy import matrix
mat = matrix([1, 2, 3])
np_asanyarray = np.asanyarray(mat)  # Retains the matrix subclass
```

```python
#Q14 a)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
temperature = np.random.randint(10, 30, size=len(date_rng))
humidity = np.random.randint(40, 80, size=len(date_rng))

df = pd.DataFrame({'Date': date_rng, 'Temperature': temperature, 'Humidity': humidity})

fig, ax1 = plt.subplots()
# b)
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
# c)
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
#Q15
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# b) Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, loc=0, scale=1)
plt.plot(x, p, 'k', linewidth=2)

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
# d) #Q16
plt.title('Histogram with PDF Overlay')

plt.show()
```

```python
# Q17
import seaborn as sns


x = np.random.rand(100) - 0.5
y = np.random.rand(100) - 0.5

quadrant = []
for i in range(len(x)):
  if x[i] >= 0 and y[i] >= 0:
    quadrant.append('Quadrant I')
  elif x[i] < 0 and y[i] >= 0:
    quadrant.append('Quadrant II')
  elif x[i] < 0 and y[i] < 0:
    quadrant.append('Quadrant III')
  else:
    quadrant.append('Quadrant IV')

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrant})

# Create the scatter plot
sns.scatterplot(x='x', y='y', hue='Quadrant', data=df)

# Add legend, labels, and title
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

```python
#Q18
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Activate Bokeh output in the notebook
output_notebook()

# Generating data for the sine wave
x = np.linspace(0, 4 * np.pi, 500)  # Generate 500 points from 0 to 4*pi
y = np.sin(x)


p = figure(height=400, width=600, title="Sine Wave Function",
           toolbar_location=None) # Changed plot_height to height and plot_width to width

# Plot the sine wave
p.line(x, y, line_width=2, color="blue", legend_label="Sine wave")

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Label axes
p.xaxis.axis_label = "x"
p.yaxis.axis_label = "sin(x)"

# Show the result
show(p)
```

```python
#Q19

import numpy as np
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.palettes import Viridis256

output_notebook()


categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))


source = ColumnDataSource(data=dict(categories=categories, values=values))


p = figure(x_range=categories, height=400, width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Color map based on values
mapper = factor_cmap(field_name='categories', palette=Viridis256, factors=categories)

# Add bars to the figure
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white', fill_color=mapper)

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the result
show(p)
```
